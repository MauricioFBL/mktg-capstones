```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
def read_data(path: str) -> pd.DataFrame:

    df = pd.read_csv(path, header=0, sep=',')

    return df


def merging_all_tables(df1: pd.DataFrame, 
                       df2: pd.DataFrame, 
                       df3: pd.DataFrame, 
                       df4: pd.DataFrame) -> pd.DataFrame:

    df = df1.merge(df2, how='inner', on='policy_type_id')
    df = df.merge(df3, how='inner', on='policy_lvl_id')
    df = df.merge(df4, how='inner', on='state_id')

    return df


def cleaning_data(df: pd.DataFrame) -> pd.DataFrame:

    df = df.drop(columns=['policy_type_id', 'policy_lvl_id', 'state_id'])

    return df
```

FUNCTIONS ORCHESTRATION

```python
transactions_df = read_data('../../data/tables/transactions.csv')
policies_levels_df = read_data('../../data/tables/policies_levels.csv')
policies_types_df = read_data('../../data/tables/policies_types.csv')
states_df = read_data('../../data/tables/states.csv')
```

```python
all_tables_df = merging_all_tables(df1=transactions_df,
                              df2=policies_types_df,
                              df3=policies_levels_df,
                              df4=states_df)
```

```python
df_ready = cleaning_data(all_tables_df)
```
